```python
%matplotlib inline
from matplotlib import style
style.use('fivethirtyeight')
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
```

# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
from sqlalchemy import distinct, desc
from sqlalchemy import text
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
base = automap_base()
# reflect the tables
base =automap_base()
base.prepare(autoload_with=engine)
```

```python
# View all of the classes that automap found
print(base.classes.keys())
```

```python
# Save references to each table
measurement = base.classes.measurement
station = base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
most_recent = session.query(func.max(measurement.date)).all()
max_date = most_recent[0][0] # using [0][0] gives the first row of and then allows you to get the date within the tuple
print(max_date)
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results Starting from the most recent data point in the database. 

#Calculate the date one year from the last date in data set.
max_date_list = max_date.split("-") #split will allow to return in date format even though sqlite does not have a date type
start_date = dt.date (int(max_date_list[0]), int(max_date_list[1]), int(max_date_list[2]))
print(start_date)
year = dt.timedelta(days=366)                     
time = start_date - year

print(time)


# Perform a query to retrieve the data and precipitation scores
precipitation_date = session.query(measurement.date, measurement.prcp).filter(measurement.date > (time)).all()

# Save the query results as a Pandas DataFrame. Explicitly set the column names
df_1 = pd.DataFrame(data=precipitation_date, columns=["Date", "Precipitation"])
df_2 = df_1.set_index("Date")

# Sort the dataframe by date
df_3 = df_2.sort_index()

# Use Pandas Plotting with Matplotlib to plot the data
df_3.plot.line(figsize=(20,10))
plt.xlabel("Date")
plt.ylabel("Inches")
plt.title("Precipitation Over A Year")
plt.tight_layout()
```

```python
# Use Pandas to calculate the summary statistics for the precipitation data
df_1.describe()
```

```python
# Design a query to calculate the total number of stations in the dataset
query = text(f"SELECT count(station) FROM Measurement")
station_count1 = session.execute(query).fetchall()
print(station_count1)
station_count2 = session.query(measurement.station)
print(station_count2.count())
station_count3 = session.query(station.station)
print(station_count3.all())
```

```python
# Design a query to find the most active stations (i.e. which stations have the most rows?)
# List the stations and their counts in descending order.
unique_station_count = session.query(measurement.station,func.count(measurement.station).label("Station Activity")).group_by(measurement.station)
sorted = unique_station_count.order_by(desc("Station Activity")).all()
for count_per_station in sorted:
    print(count_per_station)
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
min_temp = session.query(func.min(measurement.tobs)).filter(measurement.station == sorted[0][0]).scalar() 
max_temp = session.query(func.max(measurement.tobs)).filter(measurement.station == sorted[0][0]).scalar() 
avg_temp = session.query(func.avg(measurement.tobs)).filter(measurement.station == sorted[0][0]).scalar()   

#session.query(func.min(measurement.tobs))constructs a query to get the minimum temperature observations (tobs) from the measurement table, 
# .filter(measurement.station == sorted[0][0]) filters the result corresponds to the first entry in sorted
# .scalar returns a single value which gives the min, max, or average

print("Min temp: ", min_temp)
print("Max temp: ", max_temp)
print("AVG temp: ", avg_temp)                                                            
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
yearly_temp = session.query(measurement.date, measurement.tobs).filter(measurement.date > time, measurement.station == sorted[0][0]).all()
yearly_temp_df = pd.DataFrame(yearly_temp)
yearly_temp_df.plot.hist(bins =12, figsize = (20,10))
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.title("Yearly Temperatures")
```
